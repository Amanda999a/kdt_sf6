```python
#pip install folium
import folium
# Map() 클래스 사용, location  - [위도, 경도], zoom_start - 확대/축소
seoul_map = folium.Map(location=[37.539705, 126.94604], zoom_start=12)

# locations = [
#         [37.539705, 126.94604], # 마포역
#         [37.497952, 127.027619], # 강남역
#         [37.571731, 127.011069] # 동대문역
# ]

# popup을 위해 리스트[{딕셔너리}] 형태로 진행
map_info = [
    {"location" : [37.539705, 126.94604], "popup": "마포역"},
    {"location" : [37.497952, 127.027619], "popup": "강남역"},
    {"location" : [37.571731, 127.011069], "popup": "동대문역"}
]

for map in map_info:
    folium.Marker(
        location = map["location"], # key로 검색
        popup = map["popup"],
        icon = folium.Icon(icon = 'arrow-down', color='blue')
    ).add_to(seoul_map)

seoul_map
```
