#### 공덕역 마커 표시하기

```python
import folium

# Map 클래스 사용, location  - [경도, 위도], zoom_start - 확대/축소
seoul_map = folium.Map(
    location=[37.54322, 126.951576], 
    zoom_start=15,
)

# Marker 클래스 - icon인자 : Icon 클래스. icon 속성 = 'star', 'arrow-up', 'arrow-down'
# Marker 매서드 - .add_to(지도 객체)
folium.Marker(
    location=[37.54322, 126.951576], 
    icon = folium.Icon(icon = 'arrow-down', color = 'green'),
    zoom_start=15
).add_to(seoul_map)


seoul_map
```

```python
import folium

# Map 클래스 사용, location  - [경도, 위도], zoom_start - 확대/축소
seoul_map = folium.Map(
    location=[37.54322, 126.951576], 
    zoom_start=15,
)

# Marker 클래스 - icon인자 : Icon 클래스. icon 속성 = 'star', 'arrow-up', 'arrow-down'
# Marker 매서드 - .add_to(지도 객체)
folium.Marker(
    location=[37.54322, 126.951576], 
    icon = folium.Icon(icon = 'arrow-down', color = 'green'),
    popup = "공덕역",
    zoom_start=15
).add_to(seoul_map)

folium.CircleMarker(
    location=[37.54322, 126.951576], 
    radius = 100, # 반지름 100 px
    color = "#ff0" #16진수 rgb 색상값
).add_to(seoul_map)


seoul_map
```
