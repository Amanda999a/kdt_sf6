#### 시각화 라이브러리 사용 - maplotlib

```python
import matplotlib.pyplot as plt
plt.title("Basic Grapth")
# plt.plot([1,2,3,4]) # X축값이 생략되면 Y축값만 사용된다
plt.plot([1, 2, 3, 4], [10, 40, 20, 30]) # 가로 X축 값, 세로Y축 값 - (1,10), (2,40)
plt.show()
```

```python
import matplotlib.pyplot as plt
# 한글 글꼴 오류 해결
plt.rc('font', family='Malgun Gothic')
plt.title("시험 성적 그래프")
plt.plot([1, 2, 3, 4], [90, 80, 85, 70])
plt.xlabel("학생번호") #X축 제목
plt.ylabel("시험점수") #Y축 제목
plt.show()
```

```python
import matplotlib.pyplot as plt

# 한글 글꼴 오류 해결
plt.rc('font', family='Malgun Gothic')
plt.title("범례 및 색상 표시")
plt.plot([10, 20, 30, 40], label="비만율", color='r') #red
plt.plot([40, 30, 20, 10], label="출생율", color='g') #green
plt.legend() # label 작성시(범례 소제목) 필수
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.title("마커 표시")
plt.plot([10, 20, 30, 40], 'b.', label="circle") # 순서 주의. 'blue'
plt.plot([40, 30, 20, 10], 'r^', label="triangle") # 순서 주의. 'red'
plt.legend()
plt.show()
```
